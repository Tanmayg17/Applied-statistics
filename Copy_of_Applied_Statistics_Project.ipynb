```python
#outlier detection using IQR
q1 = numerical_cols.quantile(0.25)
q3 = numerical_cols.quantile(0.75)
IQR = q3-q1
Lowerbound = q1 - 1.5 * IQR
Upperbound = q3 + 1.5 * IQR
outliers_iqr = ((numerical_cols < Lowerbound) | (numerical_cols > Upperbound)).any(axis=1)
```

```python
#outliers detections using zscore
from scipy.stats import zscore
z_score = numerical_cols.apply(zscore)
threshold = 3
outliers = (np.abs(z_score) > threshold).any(axis=1)
```

**Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**

Response (target) - 1 if customer accepted the offer in the last campaign, 0 otherwise
it follows a Bernoulli distribution.



```python
#1 if customer accepted the offer in the last campaign, 0 otherwise
accepted_customers = df['Response'].sum()
accepted_customers
```

```python
#Bernoulli Distribution for Response(target)
total_customers = len(df)
probability_acceptance = accepted_customers/total_customers
print("Probability of acceptance:", probability_acceptance)

# Calculate the expected value
expected_value = probability_acceptance # the expected value of a Bernoulli distribution is simply the probability of success
print('Expected Value is:', expected_value)
```

Complain-binary outcome (1 if the customer complained, 0 otherwise), follows a Bernoulli distribution.

```python
complaints= df['Complain'].sum()
complaints
```

```python
#probability of complain
probability_complain = complaints / total_customers
print('probability of complain:',probability_complain)
```

Let try to see distribution of income we will use normal distribution

```python
#normal distribution of income
from scipy.stats import norm
mean = df['Income'].mean()
std = df['Income'].std()

#lets assign threshold value to see normal distribution upto that
threshold = 50000
probability = norm.cdf(threshold, mean, std)
expected_value = mean

#print
print('Probability of income upto 50000:', probability )
print('Expected value of income:', expected_value)
```

NumDealsPurchases, NumCatalogPurchases, NumStorePurchases, NumWebPurchases, NumWebVisitsMonth - These variables represent data over period of time or model count data we will use poisson distribution


```python
from scipy.stats import poisson
#list variables
variables = ['NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases','NumWebVisitsMonth']

for i in variables:
  avg = df[i].mean()
  expected_value = avg

  #calculate poission distribution upto 5 purchases
  probabilities = [poisson.pmf(i,avg) for i in range(6)]
  print(f"Variable: {i}")
  print("Expected value:", expected_value)
  print("Probability of number of purchases:")
  for i, prob in enumerate(probabilities):
    print(f"  {i}: {prob}")
    print()
```

**MntFishProducts, MntMeatProducts, MntFruits, MntSweetProducts, MntWines, MntGoldProds: These variables represent monetary amounts spent on different product categories in last 2 years,gamma distribution.**

```python
from scipy.stats import gamma

# List of expenditure variables
expenditure_vars = ['MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds']

for var in expenditure_vars:
    # Fit gamma distribution to the data (maximum likelihood estimation)
    shape, loc, scale = gamma.fit(df[var])

    print(f"Variable: {var}")
    print("Estimated Shape Parameter:", shape)
    print("Estimated Scale Parameter:", scale)

    # Calculate expected expenditure (mean of the fitted distribution)
    expected_expenditure = shape * scale
    print("Expected Expenditure:", expected_expenditure)

    # Calculate probability of expenditure values in a range
    min_expenditure = df[var].min()
    max_expenditure = df[var].max()
    expenditure_range = np.linspace(min_expenditure, max_expenditure, 10)  # Adjust the number of points as needed
    probabilities = gamma.pdf(expenditure_range, shape, loc=loc, scale=scale)

    # Print the probabilities for the range of expenditure values
    for i, value in enumerate(expenditure_range):
        print(f"Probability of {value}: {probabilities[i]}")

    print()
```
